```python
import numpy as np
import pandas as pd
import re
```

```python
df=pd.read_csv("/kaggle/input/job-description-dataset/job_descriptions.csv")
```

```python
df.columns
```

```python
df = df.drop_duplicates()
df.reset_index(drop=True, inplace=True)
```

```python
#CLEAN COMPANY SIZE
df['Company Size'] = pd.to_numeric(df['Company Size'], errors='coerce')
```

```python
#extract_experience
def extract_experience(val):
    if pd.isna(val):
        return np.nan
    match = re.search(r'\d+', str(val))
    return int(match.group()) if match else np.nan

df['Experience_Number'] = df['Experience'].apply(extract_experience)
```

```python
df['Experience_Even'] = df['Experience_Number'].apply(
    lambda x: True if pd.notna(x) and x % 2 == 0 else False
)
```

```python
#job title
df['Valid_JobTitle'] = df['Job Title'].apply(
    lambda x: True if isinstance(x, str) and len(x.split()) == 1 and len(x) < 10 else False
)
```

```python
# country filter
df['Country_Not_ABCD'] = df['Country'].apply(
    lambda x: False if isinstance(x, str) and x[0].upper() in ['A','B','C','D'] else True
)
```

```python
#job posting date 
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'], errors='coerce')

df['Posting_Month_Odd'] = df['Job Posting Date'].dt.month.apply(
    lambda x: True if pd.notna(x) and x % 2 != 0 else False
)
```

```python
# SHOW_3PM_5PM
df['Show_3PM_5PM'] = True
```

```python
#final check
df[['Experience', 'Experience_Number', 'Experience_Even']].head(10)
```

```python
df.columns
```

```python
import re
import numpy as np

def clean_salary(val):
    if pd.isna(val):
        return np.nan
    
    nums = re.findall(r'\d+', str(val))
    
    if len(nums) == 0:
        return np.nan
    
    # take first number as base salary (in thousands)
    return int(nums[0]) * 1000

df['Salary Range'] = df['Salary Range'].apply(clean_salary)
```

```python
df['Salary Range'].describe()
```

```python
#filling missing values
df['Salary Range'] = df['Salary Range'].fillna(df['Salary Range'].median())
```

```python
df.to_csv("cleaned_job_data.csv", index=False)
```
